```python
import thinkstats2
import nsfg
import matplotlib
import thinkplot
hist = thinkstats2.Hist([1, 2, 2, 3, 5])
hist
```

```python
hist.Freq(3)
```

```python
for val in sorted(hist.Values()):
    print(val, hist.Freq(val))
```

```python
preg = nsfg.ReadFemPreg()
live = preg[preg.outcome == 1]

matplotlib.pyplot.hist(live.birthwgt_lb.dropna())
matplotlib.pyplot.show()
```

```python
hist = thinkstats2.Hist(live.birthwgt_lb, label='birthwgt_lb')
thinkplot.Hist(hist)
thinkplot.Show(xlabel='pounds', ylabel='frequency')
```

```python
first_babies =  live[live.birthord == 1]
others_babies = live[live.birthord != 1]

matplotlib.pyplot.hist(first_babies.prglngth.dropna())
matplotlib.pyplot.hist(others_babies.prglngth.dropna())
matplotlib.pyplot.show()
```
